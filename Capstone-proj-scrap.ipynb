## Web scraping Qiyas National Center for Assessment website, for schools ranking based on average students grades in both exams Qudrat and Tahseli.

- https://www.qiyas.sa

```python
# impor required libraries:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
from selenium import webdriver
```

```python


def scraping_scores(test, sex, spec, pages = None):

    df = pd.DataFrame(columns=['الترتيب العام',
                 'الرقم الوزاري',
                 'اسم المدرسة',
                 'المدينة',
                 'المنطقة',
                 'نوع المدرسة',
                 'التصنيف',
                 'المتوسط',
                'نوع الاختبار',
                'الجنس',
                'المسار'])

    driver = webdriver.Chrome(executable_path='./chromedriver/chromedriver')
    # request the url data
    driver.get('https://www.qiyas.sa/ar/Statistics-Data/Schools/Pages/DetResults.aspx')
    
    elem_1 = driver.find_element_by_id(test)
    elem_1.click()
    
    elem_2 = driver.find_element_by_id(sex)
    elem_2.click()
        
    elem_3 = driver.find_element_by_id(spec)
    elem_3.click()

    show = driver.find_element_by_id("ctl00_ctl37_g_70814aa2_183d_441e_9095_82a1744d5849_ctl00_btnSubmit")
    show.click()

    HTML = driver.page_source
    soup = BeautifulSoup(HTML, "html.parser")

    TEST = soup.find("label", {"for": test}).text
    SEX = soup.find("label", {"for": sex}).text
    SPEC = soup.find("label", {"for": spec}).text

    sleep(2)
    status = True
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    sleep(2)
    page = 1
    while status: 
        # sleep 3 second (wait)
        sleep(2)
        #driver.implicitly_wait(2)
        # catch data
        driver.implicitly_wait(8) # seconds

        HTML = driver.page_source
        soup = BeautifulSoup(HTML, "html.parser")
        table = soup.find("div", {"class": "row photoalbumList"})
        rows = table.find_all("tr")

        for row in rows[2:] :
              df.loc[len(df)] = [x.text.strip() for x in row.find_all("td")] + [TEST, SEX, SPEC]
        
        if pages == page:
            break

        try:
            elem = driver.find_element_by_class_name('Next')
            elem.click()
            page +=1
        except:
            sleep(1)
        try:
            soup.find("a", {"class", "aspNetDisabled Next"}).text
            status = False
        except:
            pass
    driver.close()  
    return df
#close the data
# driver.close()
```
